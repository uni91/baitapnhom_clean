## 1. Phân tích đơn biến (Univariate Analysis) là gì?
**Phân tích đơn biến** là quá trình xem xét và mô tả **một biến duy nhất** trong tập dữ liệu.  
Mục tiêu chính là hiểu rõ **đặc điểm phân phối, xu hướng trung tâm và mức độ biến thiên** của biến đó.

Ví dụ:
- Với biến **“Tuổi khách hàng”**, ta có thể dùng histogram hoặc boxplot để xem phân phối tuổi.
- Với biến **“Giới tính”**, ta có thể dùng biểu đồ cột (bar chart) để xem tỷ lệ nam/nữ.

### Phân biệt với Phân tích hai biến (Bivariate Analysis)
| Tiêu chí | Phân tích đơn biến | Phân tích hai biến |
|-----------|--------------------|--------------------|
| Số lượng biến xem xét | 1 | 2 |
| Mục tiêu | Hiểu đặc tính của từng biến độc lập | Tìm hiểu mối quan hệ giữa hai biến |
| Kỹ thuật phổ biến | Histogram, Boxplot, Pie chart | Scatter plot, Crosstab, Heatmap, Correlation matrix |
| Ứng dụng | Kiểm tra phân phối, ngoại lệ | Xác định tương quan, mối liên hệ hoặc xu hướng giữa hai biến |

*Tóm lại:*  
- Phân tích **đơn biến** giúp hiểu sâu **từng biến riêng lẻ**.  
- Phân tích **hai biến** giúp hiểu **cách các biến liên hệ hoặc ảnh hưởng lẫn nhau**.


## 2. Các thước đo thống kê thường dùng trong phân tích đơn biến

### Thước đo trung tâm (Central Tendency)
| Tên | Ký hiệu | Ý nghĩa |
|------|----------|----------|
| **Trung bình (Mean)** | $\bar{x}$ | Giá trị trung bình cộng của biến |
| **Trung vị (Median)** | — | Giá trị nằm giữa khi sắp xếp dữ liệu theo thứ tự |
| **Mode (Mode)** | — | Giá trị xuất hiện thường xuyên nhất |

### Thước đo phân tán (Dispersion)
| Tên | Công thức cơ bản | Ý nghĩa |
|------|------------------|----------|
| **Độ lệch chuẩn (Standard Deviation)** | $\sqrt{\frac{\sum (x_i - \bar{x})^2}{n}}$ | Mức độ phân tán quanh trung bình |
| **Phương sai (Variance)** | $\sigma^2$ | Bình phương độ lệch chuẩn |
| **Khoảng giá trị (Range)** | Max - Min | Mức chênh lệch giữa giá trị lớn nhất và nhỏ nhất |
| **IQR (Interquartile Range)** | Q3 - Q1 | Độ rộng của vùng 50% dữ liệu giữa hai tứ phân vị |

*Ứng dụng:*  Các thước đo này giúp đánh giá xem dữ liệu có **phân tán rộng hay tập trung**, có **bị lệch** hay **có ngoại lệ** hay không.

## 3. Xác định mối quan hệ giữa hai biến trong phân tích hai biến

### Các loại mối quan hệ có thể có:
1. **Tương quan (Correlation):**  
   - Đo lường **mức độ và hướng** liên hệ tuyến tính giữa hai biến số.  
   - Dùng hệ số **Pearson (r)** hoặc **Spearman (ρ)**.
   - Ví dụ: thu nhập và chi tiêu có thể có tương quan dương (r > 0).

2. **Mối quan hệ nhân quả (Causation):**  
   - Một biến **ảnh hưởng hoặc quyết định** sự thay đổi của biến khác.  
   - Ví dụ: chi phí quảng cáo tăng → doanh thu tăng.  
*Lưu ý:* Tương quan **không đồng nghĩa với** nhân quả.

3. **Mối quan hệ giữa biến phân loại và biến số:**  
   - Dùng **boxplot** hoặc **violin plot** để so sánh phân phối.  
   - Dùng **ANOVA** hoặc **t-test** để kiểm tra sự khác biệt giữa các nhóm.

4. **Mối quan hệ giữa hai biến phân loại:**  
   - Dùng **bảng chéo (crosstab)** hoặc **heatmap**.  
   - Kiểm định bằng **Chi-square test** để xem chúng có độc lập hay không.
###  Ví dụ minh họa
| Loại biến | Mục tiêu | Phương pháp/biểu đồ gợi ý |
|------------|-----------|---------------------------|
| 2 biến số | Kiểm tra tương quan | Scatter plot, Hệ số Pearson/Spearman |
| Biến số & Biến phân loại | So sánh trung bình giữa các nhóm | Boxplot, ANOVA |
| 2 biến phân loại | So sánh tần suất | Bar chart, Crosstab, Chi-square |

### 4. Sự khác biệt giữa tương quan (Correlation) và hiệp biến (Covariance)

***a. Hiệp biến (Covariance):*** 
- Đo **mức độ cùng biến thiên** của hai biến.  
- Công thức:  
  \[
  \text{Cov}(X, Y) = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{n - 1}
  \]  
- **Ý nghĩa:**  
  - Nếu **Cov > 0** → hai biến **tăng/giảm cùng chiều**.  
  - Nếu **Cov < 0** → hai biến **biến thiên ngược chiều**.  
  - Nếu **Cov = 0** → **không có mối quan hệ tuyến tính**.  
- **Hạn chế:** Giá trị phụ thuộc vào **đơn vị đo** của biến → khó so sánh giữa các tập dữ liệu khác nhau.

***b.Tương quan (Correlation):***
- Là **phiên bản chuẩn hóa của hiệp biến**, có giá trị trong khoảng từ **-1 đến +1**.  
- Công thức:  
  \[
  r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
  \]  
- **Ý nghĩa:**  
  - **r = +1:** Tương quan tuyến tính dương hoàn hảo.  
  - **r = -1:** Tương quan tuyến tính âm hoàn hảo.  
  - **r = 0:** Không có mối tương quan tuyến tính.  
- **Ưu điểm:** Không phụ thuộc đơn vị đo → có thể so sánh giữa các dữ liệu khác nhau.

***Kết luận:***
- **Covariance** cho biết hướng mối quan hệ.  
- **Correlation** cho biết cả **hướng và mức độ chặt chẽ** của mối quan hệ giữa hai biến.


### 5. Khi nào nên sử dụng biểu đồ trực quan hóa trong phân tích đơn biến và hai biến

***a. Phân tích đơn biến (Univariate Analysis):***
- **Mục tiêu:** Tìm hiểu đặc điểm của **một biến duy nhất** (ví dụ: phân bố, trung bình, độ lệch, ngoại lệ).  
- **Biểu đồ thường dùng:**  
  - **Histogram:** Hiển thị phân bố tần suất của dữ liệu liên tục.  
  - **Boxplot:** Phát hiện giá trị ngoại lai và mức độ phân tán.  
  - **Bar chart:** So sánh tần suất giữa các danh mục.  
- **Khi sử dụng:**  
  - Khi muốn hiểu rõ **phân bố và đặc tính riêng lẻ** của một biến trước khi phân tích mối quan hệ với biến khác.  
  - Dùng trong bước **khám phá dữ liệu (EDA)** ban đầu.

***b. Phân tích hai biến (Bivariate Analysis):***
- **Mục tiêu:** Xem xét **mối quan hệ giữa hai biến** (định lượng hoặc định tính).  
- **Biểu đồ thường dùng:**  
  - **Scatter plot:** Phân tích mối tương quan giữa hai biến liên tục.  
  - **Heatmap:** Hiển thị cường độ tương quan giữa nhiều cặp biến.  
  - **Grouped bar chart / Boxplot:** So sánh phân bố giữa các nhóm.  
  - **Line chart:** Thể hiện xu hướng theo thời gian giữa hai biến.  
- **Khi sử dụng:**  
  - Khi cần đánh giá **mức độ và hướng mối quan hệ** giữa hai biến.  
  - Thường được dùng trong bước **phân tích nguyên nhân – kết quả hoặc kiểm định giả thuyết**.

***Kết luận:***
- **Đơn biến**: Dùng để hiểu rõ đặc điểm từng biến riêng lẻ.  
- **Hai biến**: Dùng để tìm hiểu **mối quan hệ** hoặc **sự phụ thuộc** giữa hai biến.


### 6. Biểu đồ Scatter Plot và Heatmap trong phân tích mối quan hệ giữa hai biến

#### a. **Scatter Plot (biểu đồ phân tán)**
- Scatter plot dùng để thể hiện **mối quan hệ giữa hai biến liên tục** bằng cách hiển thị các điểm dữ liệu trên hệ trục tọa độ.  
- Mỗi điểm biểu diễn giá trị của một cặp biến (X, Y).  
- Biểu đồ này giúp xác định **xu hướng (trend)**, **mối tương quan (correlation)**, hoặc **các giá trị ngoại lai (outliers)** trong dữ liệu.

**Ý nghĩa:**
- Nếu các điểm có xu hướng đi lên → mối tương quan dương.  
- Nếu các điểm có xu hướng đi xuống → mối tương quan âm.  
- Nếu các điểm phân tán ngẫu nhiên → không có mối tương quan rõ ràng.

**Ứng dụng:**
- Phân tích mối quan hệ giữa thu nhập và chi tiêu, thời gian học và điểm số, tuổi và hành vi mua hàng, v.v.

#### b. **Heatmap (bản đồ nhiệt)**
- Heatmap là dạng biểu đồ ma trận dùng để hiển thị **mức độ tương quan** giữa nhiều biến trong cùng một dataset.  
- Mỗi ô trong heatmap biểu thị hệ số tương quan giữa hai biến, với **màu sắc thể hiện cường độ và chiều hướng tương quan**.

**Ý nghĩa màu sắc:**
- Màu đậm hoặc nóng (đỏ) → tương quan dương mạnh.  
- Màu lạnh (xanh) → tương quan âm mạnh.  
- Màu trung tính → mối quan hệ yếu hoặc không đáng kể.

**Ứng dụng:**
- Phân tích mối quan hệ giữa các biến kinh tế, hành vi khách hàng, hoặc đặc trưng trong mô hình học máy.  
- Dễ dàng nhận diện các cặp biến có tương quan cao → phục vụ cho việc chọn đặc trưng (feature selection).

#### **Tóm lại:**
| Loại biểu đồ | Mục đích chính | Số biến | Ý nghĩa |
|---------------|----------------|----------|----------|
| **Scatter Plot** | Quan sát mối quan hệ giữa hai biến cụ thể | 2 | Thể hiện xu hướng và độ phân tán |
| **Heatmap** | Phân tích tổng thể mối tương quan giữa nhiều biến | ≥2 | Thể hiện mức độ tương quan qua màu sắc |


### 7. Trực quan hóa mối quan hệ giữa biến số và biến phân loại bằng **Boxplot** hoặc **Violin Plot**

#### a. **Boxplot (biểu đồ hộp)**
- **Mục đích:** Thể hiện **phân bố và mức độ phân tán** của một biến số (numerical variable) theo các nhóm của một biến phân loại (categorical variable).  
- **Cấu trúc biểu đồ:**
  - Đường giữa hộp: giá trị **trung vị (median)**.  
  - Mép hộp: **Q1 (25%) và Q3 (75%)**, biểu thị phạm vi chứa 50% dữ liệu.  
  - "Râu" (whiskers): phạm vi dữ liệu không phải ngoại lai.  
  - Các điểm ngoài râu: **outliers**.  

**Ý nghĩa:**
- So sánh sự khác biệt về phân bố giữa các nhóm.  
- Dễ phát hiện nhóm có giá trị trung vị cao hơn hoặc biến động lớn hơn.  

**Ví dụ ứng dụng:**
- So sánh **mức thu nhập (Income)** giữa các **nhóm trình độ học vấn (Education)**.  
- Đánh giá **điểm trung bình sinh viên** giữa **các ngành học khác nhau**.

#### b. **Violin Plot**
- **Mục đích:** Cũng hiển thị phân bố của biến số theo nhóm phân loại, nhưng kết hợp giữa **boxplot và biểu đồ mật độ (kernel density)**.  
- **Ưu điểm:**  
  - Cho thấy **hình dạng phân bố dữ liệu** (đối xứng hay lệch).  
  - Thích hợp khi muốn so sánh **cả trung vị lẫn mật độ dữ liệu**.  
**Ý nghĩa:**
- Nếu “violin” hẹp → dữ liệu tập trung nhiều tại một giá trị.  
- Nếu “violin” rộng → dữ liệu phân tán mạnh.  
- Dễ nhận biết các nhóm có phân bố tương tự hoặc khác biệt rõ rệt.

#### **So sánh nhanh**

| Đặc điểm | **Boxplot** | **Violin Plot** |
|-----------|--------------|-----------------|
| Thể hiện trung vị |  Có |  Có |
| Thể hiện phân bố (mật độ) |  Không | Có |
| Nhận diện ngoại lai | Dễ |  Có thể |
| Độ chi tiết phân bố | Trung bình | Cao |
| Khi nên dùng | So sánh tổng quan giữa các nhóm | Phân tích chi tiết hình dạng phân bố |
