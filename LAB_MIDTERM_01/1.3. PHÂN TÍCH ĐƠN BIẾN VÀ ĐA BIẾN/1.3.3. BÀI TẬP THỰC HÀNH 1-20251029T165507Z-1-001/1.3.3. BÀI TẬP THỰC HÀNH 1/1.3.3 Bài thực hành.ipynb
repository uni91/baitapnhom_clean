```python
import numpy as np
# Vá lỗi SweetViz gọi VisibleDeprecationWarning không còn trong NumPy 2.x
import warnings
if not hasattr(np, "VisibleDeprecationWarning"):
    class VisibleDeprecationWarning(DeprecationWarning):
        pass
    np.VisibleDeprecationWarning = VisibleDeprecationWarning

import pandas as pd
import sweetviz as sv

# Đọc dữ liệu
df = pd.read_csv("marketing_campaign_clean.csv")

# Tạo báo cáo
report = sv.analyze(
    source=df,
    target_feat=None,          # hoặc "Response" nếu bạn có cột mục tiêu
    feat_cfg=None,
    pairwise_analysis='auto'
)

# Xuất HTML
report.show_html("marketing_sweetviz_report.html")


```
