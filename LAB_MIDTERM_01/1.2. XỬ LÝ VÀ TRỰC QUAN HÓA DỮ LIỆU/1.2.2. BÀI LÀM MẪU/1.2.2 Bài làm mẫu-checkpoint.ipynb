```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'  # hỗ trợ Unicode
matplotlib.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")
sns.set_palette("pastel")
sns.set(font_scale=1.2)
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#C2C2F0', 
          '#FFB6C1', '#FF6666', '#99FFCC', '#FFD700', '#87CEFA']
plt.figure(figsize=(12, 6))
data = houseprices_sorted[0:10]
sns.barplot(data=data, x='Zip', y='Price', palette=colors)
plt.title("Top 10 Areas with the highest house prices", fontsize=18)
plt.xlabel("Zip code", fontsize=14)
plt.ylabel("House prices in millions", fontsize=14)
plt.show()
plt.figure(figsize=(12, 6))
data = houseprices_sorted[0:10]
ax = sns.barplot(data=data, x='Zip', y='Price', palette=colors)
ax.set_xlabel('Zip code', fontsize=15)
ax.set_ylabel('House prices in millions', fontsize=15)
ax.set_title('Top 10 Areas with the highest house prices', fontsize=20)
plt.show()
fig, ax = plt.subplots(1, 2, figsize=(30, 12))
data = houseprices_sorted[0:10]
sns.set(font_scale=2.0)
sns.barplot(data=data, x='Zip', y='Price', palette=colors, ax=ax[0])
ax[0].set_xlabel('Zip code', fontsize=18)
ax[0].set_ylabel('House prices in millions', fontsize=18)
ax[0].set_title('Top 10 Areas with the highest house prices', fontsize=20)
sns.barplot(data=data, x='Zip', y='PriceperSqm', palette=colors, ax=ax[1])
ax[1].set_xlabel('Zip code', fontsize=18)
ax[1].set_ylabel('House prices per sqm', fontsize=18)
ax[1].set_title('Top 10 Areas with the highest price per sqm', fontsize=20)

plt.tight_layout()
plt.show()
```
