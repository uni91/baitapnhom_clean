# 1. Chuẩn bị dữ liệu cho trực quan hóa dữ liệu

```python
import pandas as pd

# Đọc dữ liệu
diabetes_data = pd.read_csv("diabetes.csv")

# Chọn vài cột tiêu biểu
diabetes_data = diabetes_data[['Age', 'BMI', 'Glucose', 'Insulin']]

# Tạo cột mới: Tỉ lệ Glucose trên BMI (tương tự Price per Sqm)
diabetes_data['Glucose_per_BMI'] = diabetes_data['Glucose'] / diabetes_data['BMI']

# Hiển thị 5 dòng đầu tiên để kiểm tra
print(diabetes_data.head())
```

# 2. Trực quan hóa dữ liệu với thư viện Matplotlib


```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
diabetes_data = pd.read_csv("diabetes.csv")

# Tạo biến mới tương tự PriceperSqm
diabetes_data['Glucose_per_BMI'] = diabetes_data['Glucose'] / diabetes_data['BMI']

# Sắp xếp theo Glucose giảm dần
diabetes_sorted = diabetes_data.sort_values('Glucose', ascending=False)

# CASE 1: BASIC
plt.figure(figsize=(12,6))
x = diabetes_sorted['Age'][0:10]
y = diabetes_sorted['Glucose'][0:10]
plt.bar(x, y)
plt.show()

# CASE 2: ADVANCED 1
plt.figure(figsize=(12,6))
plt.bar(x, y)
plt.title('Top 10 Ages with the Highest Glucose Levels', fontsize=15)
plt.xlabel('Age', fontsize=12)
plt.xticks(fontsize=10)
plt.ylabel('Glucose Level', fontsize=12)
plt.yticks(fontsize=10)
plt.show()

# CASE 3: ADVANCED 2
fig, ax = plt.subplots(figsize=(40,18))
x = diabetes_sorted['Age'][0:10]
y = diabetes_sorted['Glucose'][0:10]
y1 = diabetes_sorted['Glucose_per_BMI'][0:10]

plt.subplot(1,2,1)
plt.bar(x, y)
plt.xticks(fontsize=17)
plt.ylabel('Glucose Level', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Ages with the Highest Glucose Levels', fontsize=25)

plt.subplot(1,2,2)
plt.bar(x, y1)
plt.xticks(fontsize=17)
plt.ylabel('Glucose per BMI', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 Ages with the Highest Glucose per BMI', fontsize=25)

plt.show()

```

# 3. Trực quan hóa dữ liệu với thư viện Seaborn


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
diabetes_data = pd.read_csv("diabetes.csv")

# Tạo biến mới tương tự PriceperSqm
diabetes_data['Glucose_per_BMI'] = diabetes_data['Glucose'] / diabetes_data['BMI']

# Sắp xếp theo Glucose giảm dần
diabetes_sorted = diabetes_data.sort_values('Glucose', ascending=False)

# CASE 1: BASIC
plt.figure(figsize=(12,6))
data = diabetes_sorted[0:10]
sns.barplot(data=data, x='Age', y='Glucose')
plt.title('Top 10 Ages with Highest Glucose Levels')
plt.show()

# CASE 2: ADVANCED 1
plt.figure(figsize=(12,6))
data = diabetes_sorted[0:10]
ax = sns.barplot(data=data, x='Age', y='Glucose')
ax.set_xlabel('Age', fontsize=15)
ax.set_ylabel('Glucose Level', fontsize=15)
ax.set_title('Top 10 Ages with the Highest Glucose Levels', fontsize=20)
plt.show()

# CASE 3: MULTIPLE PERSPECTIVES
fig, ax = plt.subplots(1, 2, figsize=(40,18))
data = diabetes_sorted[0:10]
sns.set(font_scale=3)

# Biểu đồ 1 – Glucose
ax1 = sns.barplot(data=data, x='Age', y='Glucose', ax=ax[0])
ax1.set_xlabel('Age')
ax1.set_ylabel('Glucose Level')
ax1.set_title('Top 10 Ages with the Highest Glucose Levels')

# Biểu đồ 2 – Glucose per BMI
ax2 = sns.barplot(data=data, x='Age', y='Glucose_per_BMI', ax=ax[1])
ax2.set_xlabel('Age')
ax2.set_ylabel('Glucose per BMI')
ax2.set_title('Top 10 Ages with the Highest Glucose per BMI')

plt.show()
```
